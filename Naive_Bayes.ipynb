```python
%matplotlib inline

from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pylab as plt

!pip install dmba
from dmba import classificationSummary, gainsChart
```

```python
ksi = pd.read_csv('KSI.csv')
```

```python
ksi.info()
```

```python
nb = ksi[['ACCLASS', 'VISIBILITY', 'LIGHT', 'RDSFCOND']]
```

```python
nb.head()
```

```python
nb.ACCLASS = nb.ACCLASS.astype('category')
nb.VISIBILITY = nb.VISIBILITY.astype('category')
nb.LIGHT = nb.LIGHT.astype('category')
nb.RDSFCOND = nb.RDSFCOND.astype('category')
```

```python
df = nb['ACCLASS'].value_counts()
print(df)
print(type(df))
```

```python
nb.replace('Property Damage Only','Non-Fatal',inplace=True)
nb.replace('Non-Fatal Injury','Non-Fatal',inplace=True)
```

```python
df = nb['LIGHT'].value_counts()
print(df)
print(type(df))
```

```python
nb.replace('Dark, artificial','Dark',inplace=True)
nb.replace('Dusk, artificial','Dusk',inplace=True)
nb.replace('Daylight, artificial','Daylight',inplace=True)
nb.replace('Dawn, artificial','Dawn',inplace=True)
```

```python
df1 = nb['LIGHT'].value_counts()
print(df)
print(type(df))
```

```python
outcome = 'ACCLASS'
predictors = ['VISIBILITY','LIGHT','RDSFCOND']

X= pd.get_dummies(ksi[predictors])
y= (ksi[outcome] == 'Fatal').astype(int)
classes= ['Non-Fatal', 'Fatal']
pd.DataFrame(y)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y, test_size=0.4, random_state=1)

nb_nb = MultinomialNB(alpha=0.1)
nb_nb.fit(X_train, y_train)

pred_prob_train = nb_nb.predict_proba(X_train)
pred_prob_valid = nb_nb.predict_proba(X_valid)

#pd.DataFrame(pred_prob_train)

y_train_pred = nb_nb.predict(X_train)
y_valid_pred = nb_nb.predict(X_valid)
```

```python
pd.DataFrame(pred_prob_train)
```

```python
index = X_train.index
indices = index.to_list()
nb.iloc[indices]
```

```python
y_train_pred = pd.DataFrame(y_train_pred)
y_train_pred.rename(columns={0:'Fatal'}, inplace=True)
y_train_pred.head()
```

```python
pred_prob_train = pd.DataFrame(pred_prob_train)
pred_prob_train.rename(columns={0:'Non-Fatal_prob', 1:'Fatal_prob'}, inplace=True)
pred_prob_train
```

```python
train_predictions = pd.concat([nb.iloc[indices].reset_index(),pred_prob_train,y_train_pred], axis=1)
```

```python
index = X_valid.index
indices = index.to_list()
nb.iloc[indices]
```

```python
y_valid_pred = pd.DataFrame(y_valid_pred)
y_valid_pred.rename(columns={0:'Fatal'}, inplace=True)
```

```python
pred_prob_valid = pd.DataFrame(pred_prob_valid)
pred_prob_valid.rename(columns={0:'Non-Fatal_prob',1:'Fatal_prob'}, inplace=True)
```

```python
valid_predictions = pd.concat([nb.iloc[indices].reset_index(),pred_prob_valid,y_valid_pred], axis=1)
valid_predictions.head()
```
